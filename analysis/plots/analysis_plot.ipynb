Codigo Python para ler os CSVs e gerar graficos

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

Leitura dos CSV

```python
bst_df = pd.read_csv('../results/bst_results.csv')
avl_df = pd.read_csv('../results/avl_results.csv')
rbt_df = pd.read_csv('../results/rbt_results.csv')
```

Configurações

```python
# Filtra os documentos de 1000 em 1000
bst_filtered = bst_df[bst_df['Documentos'] % 1000 == 0]
avl_filtered = avl_df[avl_df['Documentos'] % 1000 == 0]
rbt_filtered = rbt_df[rbt_df['Documentos'] % 1000 == 0]

x = np.array(bst_filtered['Documentos'])

# Cores
bst = '#5BC0EB'
avl = '#FF7F11'
rbt = '#9BC53D'
```

## Plots dos Gráficos

### Análise da base de dados

##### Gráfico 1 - Palavras por Documento

```python
#####  Crescimento das Palavras com os Documentos  #####

plt.style.use('seaborn-v0_8')
fig, (ax0, ax1) = plt.subplots(1, 2, figsize=(14, 5))
plt.style.use('default')


#####  Total  #####
ax0.plot(x, bst_filtered['NumeroPalavrasTotais'], marker='o', color="#6B0000EB")

ax0.set_title('Palavras Totais')
ax0.set_xlabel('Número de Documentos')
ax0.set_ylabel('Número de Palavras Totais')
ax0.set_xticks(x)
ax0.grid(alpha=0.5)


#####  Médio  #####
ax1.plot(x, bst_filtered['NumeroPalavrasDistintas'], marker='o', color="#6B0000EB")

ax1.set_title('Palvras Distintas')
ax1.set_xlabel('Número de Documentos')
ax1.set_ylabel('Número de Palavras Distintas')
ax1.set_xticks(x)
ax1.grid(alpha=0.5)


plt.subplots_adjust(top=0.85)
fig.suptitle('Crescimento das Palavras com o Número de Documentos\n', fontsize=15)

#####  SALVAR IMAGEM  #####
plt.savefig('palavras_docs.png', dpi=600)
```
